```python
import pandas as pd
import os
import json
```

```python
matches_l = []
```

```python
for file_name in os.listdir('./processed_data/matches/'):
    if not os.path.isfile(os.path.join('./processed_data/matches/', file_name)):
        print(file_name, "not a file")
        continue
    with open(os.path.join('./processed_data/matches/', file_name)) as f:
        try:
            s = json.load(f)
            matches_l.append(s)
        except UnicodeDecodeError:
            print(file_name)
```

```python
matches = pd.DataFrame.from_records(matches_l)
```

```python
matches.head()
```

```python
matches.to_csv('matches.csv', index=False)
```

```python
balls_l = []
```

```python
def process_innings_dict(inning_info: dict, match_id: str,match_inning:str):
    balls_list = []
    for ball_info in inning_info:
        for key, value in ball_info.items():
            inning, ball = key.split(".")
            runs = value.pop("runs")
            value.update(match_id=match_id, inning=inning, ball=ball,
                         runs_batsman=runs["batsman"], 
                         runs_extras=runs["extras"], runs_total=runs["total"])
            balls_list.append(value)
    return balls_list
```

```python
for file_name in os.listdir("./processed_data/innings/"):
    if not os.path.isfile(os.path.join('./processed_data/innings/', file_name)):
        print(file_name, "not a file")
        continue
    with open(os.path.join('./processed_data/innings/', file_name)) as f:
        try:
            inning_info = json.load(f)
            match_id = file_name.split("_")[0]
            match_inning = file_name.split("_")[1]
            balls_list = process_innings_dict(inning_info, match_id,match_inning)
            balls_l.extend(balls_list)
        except json.JSONDecodeError:
            print(file_name)
```

```python
balls = pd.DataFrame(balls_l)
```

```python
balls.head()
```

```python
balls.to_csv("balls.csv", index=False)
```
